### Text Pre-processing on Train Data


```python
# The first pre-processing step we’ll do is transform all sentences into lower case and create a new column sentence_lower.

df_train["sentence_lower"] = df_train["Sentence"].str.lower()
df_train.head()
```

```python
# Remove punctuations

import string

PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    #custom function to remove the punctuation
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

# Passing the function to 'Sentence' and storing into new column 'sentence_punc'

df_train["sentence_punc"] = df_train["Sentence"].apply(lambda text: remove_punctuation(text))
df_train.head()
```

```python
# remove stop words

from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

# Passing the function to 'sentence_punc' and storing into new column 'sentence_stopwords'
df_train["sentence_stopwords"] = df_train["sentence_punc"].apply(lambda text: remove_stopwords(text))
df_train.head()
```

```python
# Removal of Frequent words

from collections import Counter
cnt = Counter()
for text in df_train["sentence_stopwords"].values:
    for word in text.split():
        cnt[word] += 1
        
cnt.most_common(10)
```

```python
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

# Passing the function to 'sentence_stopwords' and storing into new column 'sentence_freqwords'
df_train["sentence_freqwords"] = df_train["sentence_stopwords"].apply(lambda text: remove_freqwords(text))
df_train.head()
```

```python
# rare words removal

n_rare_words = 10
RAREWORDS = set([w for (w, wc) in cnt.most_common()[:-n_rare_words-1:-1]])
def remove_rarewords(text):
    """custom function to remove the rare words"""
    return " ".join([word for word in str(text).split() if word not in RAREWORDS])

# Passing the function to 'sentence_freqwords' and storing into new column 'sentence_rarewords'
df_train["sentence_rarewords"] = df_train["sentence_freqwords"].apply(lambda text: remove_rarewords(text))
df_train.head()
```

```python
# Stemming

from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

# Passing the function to 'sentence_rarewords' and storing into new column 'sentence_stemmed'

df_train["sentence_stemmed"] = df_train["sentence_rarewords"].apply(lambda text: stem_words(text))
df_train.head()

```

###  Modeling

```python
# split data into X and y
X=df_train['sentence_stemmed']
y=df_train['Polarity']
```

```python
X
```

```python
y
```

```python
# split data into train and test

from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=123) # 70% train 30% test

print('Training Data :', X_train.shape)
print('Testing Data : ', X_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X_train_cv = cv.fit_transform(X_train)
X_train_cv.shape
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train_cv, y_train)
```

```python
from sklearn import metrics

# transform X_test using CV
X_test_cv = cv.transform(X_test)

# Create a prediction set:
predictions = lr.predict(X_test_cv)
```

```python
# classification report
print(metrics.classification_report(y_test,predictions))
```

```python
# Check AUC
print(metrics.roc_auc_score(y_test,lr.predict_proba(X_test_cv)[:, 1]))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train_cv, y_train)
```

```python
# predictions set

predictions_rf = rf.predict(cv.transform(X_test))
```

```python
# classification report
print(metrics.classification_report(y_test,predictions_rf))
```

```python
# Check AUC
print(metrics.roc_auc_score(y_test,rf.predict_proba(X_test_cv)[:, 1]))
```
